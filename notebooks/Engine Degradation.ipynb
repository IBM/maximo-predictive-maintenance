```python
# pass in "True" bool as second arg if we want to plot normalized data
def plot_machine(number, norm=False):
    machine = train.loc[train['Unit Number'] == number]
    plt.subplots(6, 4, figsize=(25,25))
    total_cycles = train.loc[train['Unit Number'] == number]['Cycles Left'].iloc[0]
    plt.plot(machine['Cycles Left'])
    plt.title("Cycles Left")
    eol_idx = machine.index[-1] - 50
    plt.axvline(x=eol_idx)
    for sensor_num in range(1,23):
        plt.axvline(x=eol_idx, color='r')
        plt.subplot(6, 4, sensor_num)
        sensor_vals = train.loc[train['Unit Number'] == 1]['Sensor' + str(sensor_num)]
        if norm:
            # TODO, is this norm calculation right?
            sensor_norm = (sensor_vals - sensor_vals.mean()) / (sensor_vals.max() - sensor_vals.min())
            eol_idx = sensor_norm.index[-1] - 50
            plt.plot(sensor_norm)
        else:
            plt.plot(machine['Sensor' + str(sensor_num)] )
        plt.title("Sensor " + str(sensor_num))
```
